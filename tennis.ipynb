```python
import pandas as pd

def find_algo(data):

    for index, row in data.iterrows():
        if row[-1] == 'Yes':
            hypothesis = row[:-1].tolist()
            print("Initial Hypothesis:", hypothesis)
            break

    for index, row in data.iterrows():
        if row[-1] == 'Yes':
            for i in range(len(hypothesis)):
                if row[i] != hypothesis[i]:
                    hypothesis[i] = "?"

    print("Final Hypothesis:", hypothesis)
```

```python
data=pd.read_csv('Tennis.csv');
find_algo(data)
```
