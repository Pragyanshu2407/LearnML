```python
x=[1,2,3,4,5]
y=[10,20,25,30,50]

plt.plot(x,y,marker='o',linestyle='--',color='b',label='Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.legend()
plt.show()
```

```python
categories=['A','B','C']
values=[10,20,30]
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar graph')
plt.bar(categories,values,color='g')
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y,color='orange',marker='x')
plt.xlabel('X Values')
plt.ylabel('Y values')
plt.title('Scatter plot Examples')
plt.show()
```

```python
labels=['Apple','Banana','Cherry','Date']
sizes=[30,20,35,15]

plt.pie(sizes,labels=labels,autopct='%1.1f%%',colors=['blue','red','green','yellow'])
plt.title('Fruit Distribution')
plt.show()
```
